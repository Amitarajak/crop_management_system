```python
soil_moisture_threshold_low = 25.00
soil_moisture_threshold_high= 50.00
temperature_threshold_high = 15.00
temperature_threshold_low = 30.00
soil_humidity_threshold_low=25.00
soil_humidity_threshold_high=60.00
Air_temperature_threshold_low=15.00
Air_temperature_threshold_high=21.00
wind_speed_threshold_low=1.00
wind_speed_threshold_high=3.00
Air_humidity_threshold_low=45.00
Air_humidity_threshold_high=60.00
wind_gust_threshold_low=5.00
wind_gust_threshold_high=20.00
pressure_threshold_low=80.00
pressure_threshold_high=115.00
ph_threshold_low=6.00
ph_threshold_high=8.00
rainfall_threshold_low=120.00
rainfall_threshold_high=180.00
Nitrogen_threshold_low=55.00
Nitrogen_threshold_high=80.00
phosphorus_threshold_low=30.00
phosphorus_threshold_high=55.00
potassium_threshold_low=35.00
potassium_threshold_high=45.00
humidity_threshold_low=35.00
humidity_threshold_high=45.00
```

```python
def assess_conditions(row):
    conditions = []
    if not (soil_moisture_threshold_low <= row['Soil Moisture'] <= soil_moisture_threshold_high):
        conditions.append('Inaccurate moisture level')
    if not (temperature_threshold_low <= row['temperature'] <= temperature_threshold_high):
        conditions.append('Inaccurate temperature level')
    if not (Air_temperature_threshold_low<= row['Air temperature (C)'] <= Air_temperature_threshold_high):
        conditions.append('Inaccurate air temperature level')
    if not ( wind_speed_threshold_low<= row['Wind speed (Km/h)'] <= wind_speed_threshold_high):
        conditions.append('Inaccurate wind speed')
    if not (Air_humidity_threshold_low<= row['Air humidity (%)'] <= Air_humidity_threshold_high):
        conditions.append('Inaccurate air humidity')
    if not (wind_gust_threshold_low <= row['Wind gust (Km/h)'] <= wind_gust_threshold_high):
        conditions.append('Inaccurate wind gust')
    if not (pressure_threshold_low<= row['Pressure (KPa)'] <= pressure_threshold_high):
        conditions.append('Inaccurate pressure')
    if not (ph_threshold_low<= row['ph'] <= ph_threshold_high):
        conditions.append('Inaccurate ph')
    if not (rainfall_threshold_low<= row['rainfall'] <= rainfall_threshold_high):
        conditions.append('Inaccurate rainfall')
    if not (Nitrogen_threshold_low<= row['N'] <= Nitrogen_threshold_high):
        conditions.append('Inaccurate nitrogen')
    if not (phosphorus_threshold_low<= row['P'] <= phosphorus_threshold_high):
        conditions.append('Inaccurate phosphorus')
    if not (potassium_threshold_low<= row['K'] <= potassium_threshold_high):
        conditions.append('Inaccurate potassium')


    if not conditions:
        return 'Perfect for growth'
    else:
        return 'Not perfect for growth: ' + ', '.join(conditions)

# Calculating conditions for each row in the DataFrame
merged_df['Conditions'] = merged_df.apply(assess_conditions, axis=1)

# Printing assessment results
print("\nAssessment Results:")
for idx, conditions in merged_df.head(100)['Conditions'].items():
    print(f"Index: {idx}, Conditions: {conditions}")
```

```python

columns_to_drop = ['Soil Moisture', 'Time', 'Air temperature (C)', 'Wind speed (Km/h)', 'Air humidity (%)', 'Wind gust (Km/h)', 'Pressure (KPa)','Conditions']


other_df = merged_df.drop(columns=columns_to_drop)


print("First few rows of the other DataFrame:")
print(other_df.head(30))
```

```python

temperature_threshold_low = 15
temperature_threshold_high = 30
ph_threshold_low=5
ph_threshold_high=8
rainfall_threshold_low=160
rainfall_threshold_high=270
Nitrogen_threshold_low=45
Nitrogen_threshold_high=95
phosphorus_threshold_low=20
phosphorus_threshold_high=60
potassium_threshold_low=30
potassium_threshold_high=55
humidity_threshold_low=40.00
humidity_threshold_high=90.00
```

```python
def assess_conditionss(row):
    conditionss = []

    if not (temperature_threshold_low <= row['temperature'] <= temperature_threshold_high):
        conditionss.append('Inaccurate temperature level')
    if not (ph_threshold_low<= row['ph'] <= ph_threshold_high):
        conditionss.append('Inaccurate ph')
    if not (rainfall_threshold_low<= row['rainfall'] <= rainfall_threshold_high):
        conditionss.append('Inaccurate rainfall')
    if not (Nitrogen_threshold_low<= row['N'] <= Nitrogen_threshold_high):
        conditionss.append('Inaccurate nitrogen')
    if not (phosphorus_threshold_low<= row['P'] <= phosphorus_threshold_high):
        conditionss.append('Inaccurate phosphorus')
    if not (potassium_threshold_low<= row['K'] <= potassium_threshold_high):
        conditionss.append('Inaccurate potassium')
    if not (humidity_threshold_low<= row['K'] <= humidity_threshold_high):
        conditionss.append('Inaccurate potassium')



    if not conditionss:
        return 'Perfect for growth'
    else:
        return 'Not perfect for growth: ' + ', '.join(conditionss)


other_df['Conditionss'] = other_df.apply(assess_conditionss, axis=1)


print("\nAssessment Results:")
for idx, conditionss in other_df.head(200)['Conditionss'].items():
    print(f"Index: {idx}, Conditions: {conditionss}")
```

```python
Y_newest=Y[['Conditionss_Perfect for growth']]
Y_newest.head(10)
```

```python
Y_newest = Y_newest.astype(int)
Y_newest.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, Y_newest, test_size=0.2, random_state=42)
```

```python
y_test
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

#  Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

#  Random Forest Regressor model
random_forest_model = RandomForestRegressor(random_state=42)
random_forest_model.fit(X_train, y_train)

# Cross-validation for Linear Regression
linear_scores = cross_val_score(linear_model, X_train, y_train, cv=5)
print("Linear Regression Cross-Validation Scores:", linear_scores)

#  Cross-validation for Random Forest
random_forest_scores = cross_val_score(random_forest_model, X_train, y_train, cv=5)
print("Random Forest Cross-Validation Scores:", random_forest_scores)

# Evaluating models on the test set
linear_score_test = linear_model.score(X_test, y_test)
random_forest_score_test = random_forest_model.score(X_test, y_test)

print("Linear Regression Test Score:", linear_score_test)
print("Random Forest Test Score:", random_forest_score_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers


model = keras.Sequential([
    # Input layer
    layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    # Hidden layer
    layers.Dense(32, activation='relu'),
    # Output layer
    layers.Dense(1, activation='sigmoid')
])


model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])


model.summary()


history = model.fit(X_train, y_train, epochs=10, batch_size=64, validation_split=0.2)


model.save("/content/saved_str.keras")
test_loss, test_acc = model.evaluate(X_test, y_test)
print('Test accuracy:', test_acc)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report


y_pred_prob = model.predict(X_test)


y_pred = (y_pred_prob > 0.5).astype(int).ravel()


conf_matrix = confusion_matrix(y_test, y_pred)


class_names = ['Perfect for growth', 'Not perfect for growth']


plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()


print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))
```

```python


import matplotlib.pyplot as plt


accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(accuracy) + 1)

# Plotting Accuracy
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs, accuracy, 'bo-', label='Training Accuracy')
plt.plot(epochs, val_accuracy, 'ro-', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Plotting Loss
plt.subplot(1, 2, 2)
plt.plot(epochs, loss, 'bo-', label='Training Loss')
plt.plot(epochs, val_loss, 'ro-', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
